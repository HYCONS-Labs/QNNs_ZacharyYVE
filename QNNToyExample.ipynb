```python
from LeastSquaresQNN import LSQNN

import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
```

```python
x = np.expand_dims(np.linspace(-2,2,100), axis=1)
y = 3*x - 2 + np.random.normal(0,0.1,(x.shape[0],1))

qnn = LSQNN()
qnn.Train(x, y)
y_pred = qnn.Eval(x)

plt.figure()
plt.scatter(x,y)
plt.plot(x,y_pred, color="red")

plt.figure()
plt.imshow(qnn.Z_matrix())
plt.colorbar()

plt.figure()
plt.imshow(qnn.Z_matrix(quadratic_params=False))
plt.colorbar()
```
